````python
# imports
import numpy as np
import pandas as pd
import seaborn as sns
import statistics as stats
from scipy.stats import skew, kurtosis

# data types
from numpy import float64
from numpy._typing._array_like import NDArray
````

### Exercise 1

We will use NumPy to obtain information to describe statistically.

- Generate an array of 100 elements following a normal distribution.
- Generate an array of 100 elements following a chi-square distribution with 3 degrees of freedom.
- Calculate the main metrics and statistical measures that best describe the two vectors.

````python
# normal distribution
normal_dist: NDArray[float64] = np.random.normal(loc=0, scale=0.1, size=100)
# calculating the values
print('Normal distribution:')
print(f'    Mean: {normal_dist.mean()}')
print(f'    Variance: {stats.variance(normal_dist)}')
print(f'    Standar deviation: {np.std(normal_dist, ddof=1)}')
print(f'    Median: {np.median(normal_dist)}')
print(f'    Mode: {stats.mode(normal_dist)}')
print(f'    Skewness: {skew(normal_dist)}')
print(f'    Skurt: {kurtosis(normal_dist)}')

# chi distribution
chi_dist: NDArray[float64] = np.random.chisquare(df=3, size=100)
print('\nChisquare distribution:')
print(f'    Mean: {chi_dist.mean()}')
print(f'    Variance: {stats.variance(chi_dist)}')
print(f'    Standar deviation: {np.std(chi_dist, ddof=1)}')
print(f'    Median: {np.median(chi_dist)}')
print(f'    Mode: {stats.mode(chi_dist)}')
print(f'    Skewness: {skew(chi_dist)}')
print(f'    Skurt: {kurtosis(chi_dist)}')
````

### Graph the distributions

````python
def graph(x:pd.DataFrame, column_name: str, kde: bool = False) -> None:
  '''Graph the function using seaborn'''
  sns.displot(data=x, x=column_name, kde=kde)
````

````python
# transform the disst to a dataframe
column_name = 'x'
x_as_df: pd.DataFrame = pd.DataFrame(data=normal_dist, columns=[column_name])
# graph them using seaborn
graph(x=x_as_df, column_name=column_name, kde=True)
````

````python
# transform the disst to a dataframe
column_name = 'x'
x_as_df: pd.DataFrame = pd.DataFrame(data=chi_dist, columns=[column_name])
# graph them using seaborn
graph(x=x_as_df, column_name=column_name, kde=True)
````
